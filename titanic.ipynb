```
from urllib.request import urlretrieve
url = "https://github.com/coolcue123/ml0223/raw/main/train.csv"
urlretrieve(url, "train.csv")
url = "https://github.com/coolcue123/ml0223/raw/main/test.csv"
urlretrieve(url, "test.csv")
```

```
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
test_df = pd.read_csv("test.csv", encoding="utf-8")
```

```
test_df
```

```
datas = pd.concat([train_df,test_df], axis=0, ignore_index=True)
datas
```

```
s = datas.isna().sum()
# Series[True / False]
s[s > 0].sort_values(ascending=False)
```

```
datas = datas.drop(["PassengerId", "Survived"], axis=1)
datas
```

```
# 數值類型:中位數
med = datas.median().drop("Pclass")
datas = datas.fillna(med)
```

```
# 類別類型:最常出現
most = datas["Embarked"].value_counts().idxmax()
datas["Embarked"] = datas["Embarked"].fillna(most)
```

```
s = datas.isna().sum()
# Series[True / False]
s[s > 0].sort_values(ascending=False)
```

```
def cabin(s):
  if pd.isna(s):
    return None
  else:
    return s[0]
datas["Cabin"] = datas["Cabin"].apply(cabin)
```

```
def name(s):
  mid = s.split(",")[-1].split(".")[0]
  mid = mid.strip()
  return mid
count = datas["Name"].apply(name).value_counts()
reserved = count[:4].index
def name2(s):
  mid = s.split(",")[-1].split(".")[0]
  mid = mid.strip()
  if mid in reserved:
    return mid
  else:
    return None
datas["Name"] = datas["Name"].apply(name2)
```

```
datas
```

```
dic = datas["Ticket"].value_counts()
def ticket(t):
  return dic[t]
datas["Ticket"] = datas["Ticket"].apply(ticket)
```

```
datas = pd.get_dummies(datas)
datas = pd.get_dummies(datas, columns=["Pclass"])
datas
```
